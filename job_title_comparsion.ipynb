```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv(r'C:\Users\Toan Nguyen\Desktop\project3\Data\ds_salaries.csv')
```

```python
print(data.head())  # Display the first few rows of the dataset
print(data.info())  # Get information about the dataset, such as column names and data types
```

```python
salary_by_job = data.groupby('job_title')['salary'].mean().sort_values(ascending=False)
print(salary_by_job)
```

```python
plt.figure(figsize=(12, 6))
salary_by_job.plot(kind='bar')
plt.xlabel('Job Title')
plt.ylabel('Mean Salary')
plt.title('Mean Salary by Job Title')
plt.xticks(rotation=90)
plt.show()
```

```python
highest_salary = data['salary'].max()
print(highest_salary)
```

```python
lowest_salary = data['salary'].min()
print(lowest_salary)
```

```python
average_salary = data['salary'].mean()
print(average_salary)
```

```python
average_salary_by_remote_work = data.groupby('remote_ratio')['salary'].mean()
print(average_salary_by_remote_work)
```

```python
plt.figure(figsize=(6, 4))
average_salary_by_remote_work.plot(kind='bar')
plt.xlabel('remote_ratio')
plt.ylabel('Average Salary')
plt.title('Average Salary by Remote Work')
plt.show()
```

```python
average_remote_ratio_by_job_title = data.groupby('job_title')['remote_ratio'].mean()
print(average_remote_ratio_by_job_title)
```

```python
plt.figure(figsize=(12, 6))
average_remote_ratio_by_job_title.plot(kind='bar')
plt.xlabel('job_title')
plt.ylabel('Average Remote Ratio')
plt.title('Average Remote Ratio by Job Title')
plt.xticks(rotation=90)
plt.show()
```
